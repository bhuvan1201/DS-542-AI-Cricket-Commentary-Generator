### Single Event

```python
from gtts import gTTS

# Your single event commentary text
text = "Dhoni finishes off in style! A magnificent strike into the crowd. India lift the World Cup after 28 years!"

# Initialize the gTTS object
tts = gTTS(text=text, lang='en')   #'en' stands for English language

# Save the generated speech to an .mp3 file
tts.save("commentary.mp3")
```

```python
from IPython.display import Audio  #to display the audio

# Play the generated mp3 file
Audio("commentary.mp3")
```

### Multiple Events

```python
from gtts import gTTS
from IPython.display import Audio, display
import os

# List of 10 famous cricket commentary lines
commentary_lines = [
    "A beautiful delivery swings in, beats the bat, and the keeper takes it cleanly. No run.",
    "A crisp cover drive, perfectly timed, races through the gap, and thats four runs!",
    "Skips down the track with a powerful hit through midwicket, and it speeds away to the boundary for four!",
]

# Create directory to store MP3s
os.makedirs("commentary_audio", exist_ok=True)

# Generate and save audio for each commentary line
for i, line in enumerate(commentary_lines, start=1):
    tts = gTTS(text=line, lang='en')
    filename = f"commentary_{i}.mp3"
    tts.save(filename)
    print(f" Saved: {filename}")

    # Optional: Play audio in notebook
    display(Audio(filename))
```
